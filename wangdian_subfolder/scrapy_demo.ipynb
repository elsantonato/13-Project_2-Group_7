```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
# get movie rate from https://www.rottentomatoes.com/m
base_url = 'https://www.rottentomatoes.com/m/'
base_url_meta = 'https://www.metacritic.com/movie/'
# add the correct User-Agent
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.77 Safari/537.36'}
```

```python
movie = 'Titanic'
movie_space = 'I Am Not a Witch'
```

```python
def get_url(base_url, movie_name):
    tag_list = movie_name.lower().split()
    
    tag = ''
    for item in tag_list:
        if(item != tag_list[-1]):
            tag = tag + str(item) + '_'
        else:
            tag = tag + str(item)
    
    url = base_url + tag
    
    return url
```

```python
movie_space_url = get_url(base_url, movie_space)
movie_space_url
```

```python
titanic_url = get_url(base_url, movie)
titanic_url
```

```python
titanic_url_meta = get_url(base_url_meta, movie)
titanic_url_meta
```

```python
def get_html(url):
    try:
        r = requests.get(url, timeout=30, headers=headers)
        r.raise_for_status
        #site encoding
        r.encoding = 'utf-8'
        return r.text
    except:
        print('ERROR')
```

```python
def get_content(url):
    html = get_html(url)
    soup = BeautifulSoup(html, 'lxml')
    data = soup.find('a', id='tomato_meter_link')
    
    return data
```

```python
get_content(movie_space_url)
```

```python
def get_content_metacritic(url):
    html = get_html(url)
    soup = BeautifulSoup(html, 'lxml')
    data = soup.find('div', id='nav_to_metascore')
    
    return data
```

```python
titanic_data_meta = get_content_metacritic(titanic_url_meta)
titanic_data_meta
```
