```python
# Import Dependencies
import requests
from pprint import pprint
```

```python
# Try scraping the-numbers.com (for Top 100 Grossing Movies)
response = requests.get("https://www.the-numbers.com/box-office-records/domestic/all-movies/cumulative/all-time")
response
import pandas as pd
df_1 = pd.read_html(response.text)[0]
df_1.head()
```

```python
df_2 = pd.read_html(response.text)[1]
df_2.head()
```

```python
df = df_1.append(df_2).reset_index(drop=True)
df.head()
```

```python
# Convert money text to decimal formatting
from decimal import Decimal
import re

# Make a copy of dataframe
df2 = df.copy()

# Loop through copy of dataframe
for index, row in df2.iterrows():
    # Convert to decimal
    money = re.match(r'\$\d+,\d+,\d+', row['DomesticBox Office']).group()
    df2 = df2.replace(row['DomesticBox Office'], Decimal(re.sub(r'[^\d.]', '', money)))
    money = re.match(r'\$\d+,\d+,\d+', row['InternationalBox Office']).group()
    df2 = df2.replace(row['InternationalBox Office'], Decimal(re.sub(r'[^\d.]', '', money)))
    money = re.match(r'\$\d+,\d+,\d+', row['WorldwideBox Office']).group()
    df2 = df2.replace(row['WorldwideBox Office'], Decimal(re.sub(r'[^\d.]', '', money)))
    
df2.head()
```

```python
df2.to_csv('../movie_base.csv', encoding='utf-8')
```
